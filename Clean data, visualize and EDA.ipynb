```python
# Check for duplicate data
df.duplicated().sum()
```

```python
# Delete duplicate data
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
# rename hotel
df.columns = ['Hotel', 'Canceled', 'LeadTime', 'ArrivalDateYear', 'ArrivalDateMonth', 'ArrivalDateWeekNumber','ArrivalDatedayOfMonth', 'StaysInWeekendNights',
              'StaysInWeekNights', 'Adults', 'Children', 'Babies', 'Meal','Country', 'MarketSegment', 'DistributionChannel','IsRepeatGuest', 'PreviousCancellations',
              'PreviousBookingsNotCanceled', 'ReservedRoomType','AssignedRoomType', 'BookingChanges', 'DepositType','agent', 'DaysInWaitingList',
              'CustomerType', 'ADR','RequiredCarParkingSpace', 'TotalOfSpecialRequest','ReservationStatues', 'ReservationStatus']

df.head()
```

```python

# Identify Continuous and Categorical Variables
def var(hotel):
    unique_list = pd.DataFrame([[i,len(hotel[i].unique())] for i in hotel.columns])
    unique_list.columns = ['name','uniques']

    total_var = set(hotel.columns)
    cat_var = set(unique_list.name[(unique_list.uniques<=12)|
                                   (unique_list.name=='Country')|
                                   (unique_list.name=='Agent')])
    con_var = total_var - cat_var

    return cat_var, con_var

cat_var, con_var = var(df)

print("Continuous Variables (",len(con_var),"):\n",con_var,'\n\n'
        "Categorical Variables(",len(cat_var),"):\n",cat_var)
```

```python
df.loc[df.LeadTime                    > 500,'LeadTime'                   ]=500
df.loc[df.DaysInWaitingList           >   0,'DaysInWaitingList'          ]=  1
df.loc[df.StaysInWeekendNights        >=  5,'StaysInWeekendNights'       ]=  5
df.loc[df.Adults                      >   4,'Adults'                     ]=  4
df.loc[df.PreviousBookingsNotCanceled >   0,'PreviousBookingsNotCanceled']=  1
df.loc[df.PreviousCancellations       >   0,'PreviousCancellations'      ]=  1
df.loc[df.StaysInWeekNights           >  10,'StaysInWeekNights'          ]= 10
df.loc[df.BookingChanges              >   5,'BookingChanges'             ]=  5


cat_var = set(list(cat_var) + ['PreviousBookingsNotCanceled','PreviousCancellations'])
con_var = set(df.columns) - cat_var

df[con_var].describe()
```

```python
df.loc[df.Babies                      >   8,'Babies'                 ] = 0
df.loc[df.RequiredCarParkingSpace     >   5,'RequiredCarParkingSpace'] = 0
df.loc[df.Children                    >   8,'Children'               ] = 0

df[cat_var].describe()
```

### **Data Preparation**

```python
#Combine children and babies together as kids
df['Kids'] = df.Children + df.Babies

#Combine total mumbers by adding kids and adults
df['TotalMembers'] = df.Kids + df.Adults
```

```python
#convert the datatypes to string
df['ArrivalDateYear'] = df['ArrivalDateYear'].astype('str')
df['ArrivalDateMonth'] = df['ArrivalDateMonth'].astype('str')
df['ArrivalDatedayOfMonth'] = df['ArrivalDatedayOfMonth'].astype('str')

df['PreviousCancellations'] = df['PreviousCancellations'].astype('str')
df['IsRepeatGuest'] = df['IsRepeatGuest'].astype('str')
```

```python
# Convert arrival date to datetime
df['Arrival Date'] = df['ArrivalDatedayOfMonth'] + '-' + df['ArrivalDateMonth'] + '-' + df['ArrivalDateYear']
df['Arrival Date'] = pd.to_datetime(df['Arrival Date'], errors='coerce')
```
